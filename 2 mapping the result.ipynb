```python
rng = pd.date_range('1/1/2010', periods=8760, freq='H')
All_areas = pd.DataFrame(index=rng)
for key in store.keys():
    All_areas[key] = store[key].sum(axis=1)
```

```python
def make_single_map(day_d, plottitle, value_unit):
    """make a simple map."""

    this_area = All_areas[day_d:day_d]
    
    savename = "HeatDemand {}".format(day_d)
    save_folder = "./MAPS/"
    mapper = Mapper(savename, folder=save_folder)
    savename = "HeatDemand {}".format(day_d)

    statgebs = []
    heat_demand = []
    statgebiet = GeoData(DATANAME)
    for stg in statgebiet.statistische_gebiete_utm:
        statgeb = str(int(stg.statgeb))
        area = "/area" + statgeb
        if area in All_areas.columns:
            statgebs.append(statgeb)
            heat_demand.append(float(All_areas[day_d:day_d][area]) / 1e9)
            mapper.addPatch(stg.wkb_geometry, zpos=1)
    
    hhb = GeoData("bezirke_new")
    mapper.addElement(hhb.bezirke_new, Multiple=True, zpos=0, faceCol="#aba390")

    mapper.makeStaticMap(
        scale=heat_demand,
        names=statgebs,
        smallprint=SOURCEDATA,
        title=plottitle,
        unit=value_unit)
    
    mapper.close()
    
    image_path = os.path.join(os.getcwd(), save_folder)
    image_path = os.path.join(image_path, "{}.png".format(savename))
    return(image_path)
```

```python
def update_file_list():
    maps = []
    for map_file in os.listdir(os.path.join(os.getcwd(), "MAPS")):
        maps.append(map_file)
    return(maps)
```

```python
def make_map(day_d):
    map_name = "HeatDemand {}".format(day_d)
    map_files = update_file_list()
    if map_name not in map_files:
        image_path = make_single_map(day_d, "Heat demand for {}".format(day_d), "GJ")
    else:
        savename = "HeatDemand {}".format(day_d)
        save_folder = "MAPS"
        image_path = os.path.join(os.getcwd(), save_folder, "{}.png".format(savename))
        image_path = os.path.join(image_path, "{}.png".format(savename))
    return(image_path)
```

```python
@interact
def showMap(month:(1,12)=1, day:(1,31)=1, hour:(0,23)=0):
    day_d = dt.datetime(2010, month, day, hour)
    image_path = make_map(day_d)
    display(Image(filename=image_path, retina=True))
```

```python
def click_handler(widget):
    if aggregator_info_value.value == "none":
        last_day_month = calendar.monthrange(2010, start_month.value)[1]
        if end_day.value > last_day_month:
            new_end_day = last_day_month
        else:
            new_end_day = end_day.value
        start_date = dt.datetime(2010, start_month.value, start_day.value, start_hour.value)
        end_date = dt.datetime(2010, end_month.value, new_end_day, end_hour.value)
        numer_maps = (end_date - start_date).seconds / 3600
    elif aggregator_info_value.value == "day":
        last_day_month = calendar.monthrange(2010, end_month.value)[1]
        if end_day.value > last_day_month:
            new_end_day = last_day_month
        else:
            new_end_day = end_day.value
        start_date = dt.date(2010, start_month.value, start_day.value)
        end_date = dt.date(2010, end_month.value, new_end_day)
        numer_maps = (end_date - start_date).days
    elif aggregator_info_value.value == "month":
        start_date = dt.date(2010, start_month.value, 1)
        end_date = dt.date(2010, end_month.value, 1)
        numer_maps = end_date.month - start_date.month + 1
        
    print("start date: {}".format(start_date))
    print("end date: {}".format(end_date))
    
    if define_file_name.value:
        this_file_name = "{} to {}".format(start_date, end_date)
    else:
        this_file_name = file_name.value
    if file_name_sufix.value:
        sufix = file_name_sufix.value
    else:
        sufix = aggregator_info_value.value
    if aggregator_info.visible:
        print("aggregate to {} using {}".format(aggregator_info_value.value, aggregator_info_how.value))
        this_file_name = "./MAPS/gif/{}_{}.gif".format(this_file_name, sufix)
    else:
        this_file_name = "./MAPS/gif/{}.gif".format(this_file_name)
    file_name.value = this_file_name
    
    if start_date > end_date:
        raise Exception("start date has to be before end date\n")
    
    progress.value = 0
    temp_folder = os.path.join(os.getcwd(), "MAPS/gif/pics/")
    for f in os.listdir(temp_folder):
        os.remove(os.path.join(temp_folder, f))
    for single_date in daterange(start_date, end_date):
        image_path = make_map(single_date)
        tmp_image_path = "{}".format(os.path.join(temp_folder, image_path.split("/")[-1]))
        copyfile(image_path, tmp_image_path)
        progress.value += 1/(numer_maps+1)
    progress.value = 1
    this_file_name = this_file_name.replace(" ", "\ ")
    call("/usr/bin/convert -resize 50% -delay 90 -loop 0 ./MAPS/gif/pics/* {:s}".format(this_file_name), shell=True)
    
    webm_file_name = this_file_name.split(".")[:-1]
    webm_file_name.append("webm")
    webm_file_name = ".".join(webm_file_name)
    call("/usr/bin/ffmpeg -i {:s} {:s}".format(this_file_name, webm_file_name), shell=True)
    print("stored as: {}".format(webm_file_name))
```

```python
def daterange(start_date, end_date):
    if aggregator_info_value.value == "none":
        for n in range(int((end_date - start_date).seconds / 3600 + 1)):
            yield start_date + dt.timedelta(0, n * 3600)
    elif aggregator_info_value.value == "day":
        for n in range(int((end_date - start_date).days + 1)):
            yield start_date + dt.timedelta(n)
    elif aggregator_info_value.value == "month":
        for month in range(start_date.month, end_date.month + 1):
            yield dt.date(2010, month, 1)
```

```python
def aggregate(to, how):
    if to == "month":
        All_areas_a = All_areas.resample('D', how=how)
    elif to == "day":
        All_areas_a = All_areas.resample('M', how=how)
    return(All_areas_a)
```

```python
def on_aggregator_toggle(name, value):
    if value:
        aggregator_info.visible = True
    else:
        aggregator_info.visible = False

def on_state_toggle(name, value):
    if value == 'month':
        month_info.visible = True
        day_info.visible = False
        hour_info.visible = False
    elif value == 'day':
        month_info.visible = True
        day_info.visible = True
        hour_info.visible = False
    else:
        month_info.visible = True
        day_info.visible = True
        hour_info.visible = True
        
def on_name_toggle(name, value):
    if value:
        name_info.visible = False
    else:
        name_info.visible = True
```

```python
form = widgets.ContainerWidget()

define_file_name = widgets.CheckboxWidget(description="Default Name:", value=True)
name_info = widgets.ContainerWidget(visible=False)
file_name = widgets.TextWidget(description="File Name:")
name_info.children = [file_name]
aggregator = widgets.CheckboxWidget(description="Aggregate:", value=False)
aggregator_info = widgets.ContainerWidget(visible=False)
file_name_sufix = widgets.TextWidget(description="Name Sufix:")
aggregator_info_value = widgets.DropdownWidget(description='Aggregation:', values=['none', 'month', 'day'])
aggregator_info_how = widgets.DropdownWidget(description='How:', values=['sum', 'mean', 'min', 'max'])

aggregator_info.children = [file_name_sufix, aggregator_info_value, aggregator_info_how]

start = widgets.ButtonWidget(description="Make Map")

month_info = widgets.ContainerWidget(visible=True)
start_month = widgets.IntSliderWidget(min=1, max=12, value=1, description="Start month")
end_month = widgets.IntSliderWidget(min=1, max=12, value=12, description="End month")
month_info.children = [start_month, end_month]

day_info = widgets.ContainerWidget(visible=True)
start_day = widgets.IntSliderWidget(min=1, max=31, value=1, description="Start day")
end_day = widgets.IntSliderWidget(min=1, max=31, value=31, description="End day")
day_info.children = [start_day, end_day]

hour_info = widgets.ContainerWidget(visible=True)
start_hour = widgets.IntSliderWidget(min=0, max=23, value=0, description="Start hour")
end_hour = widgets.IntSliderWidget(min=0, max=23, value=23, description="End hour")
hour_info.children = [start_hour, end_hour]

progress = widgets.FloatProgressWidget(max=1, min=0, description="Progress")

form.children = [define_file_name, name_info, aggregator, aggregator_info,
                 month_info, day_info, hour_info,
                 start, progress]
display(form)

aggregator.on_trait_change(on_aggregator_toggle, 'value')
define_file_name.on_trait_change(on_name_toggle, 'value')
aggregator_info_value.on_trait_change(on_state_toggle, 'value')
start.on_click(click_handler)
my_file_name = os.path.join(os.getcwd(), file_name.value[2:])
```
